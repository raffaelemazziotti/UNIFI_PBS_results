```python
import pandas as pd

# Load Excel files
mod1 = pd.read_excel('data/Correzione parziale modulo 1.xlsx')
mod2 = pd.read_excel('data/Correzione parziale modulo 2.xlsx')

# Merge
df_merged = pd.merge(
    mod1,
    mod2,
    on="ID number",
    how="outer",
    suffixes=("_mod1", "_mod2")
)

# Desired column order
col_order = [
    'ID number',
    'Results_mod1', 'Results_mod2',
    'Voto 1_mod1', 'Voto 2_mod1', 'Voto 3_mod1', 'Voto 4_mod1',
    'Voto 5_mod1', 'Voto 6_mod1', 'Voto 7_mod1',
    'Voto 8_mod1', 'Voto 9_mod1', 'Voto 10_mod1', 'Voto 11_mod1', 'Voto 12_mod1',
    'Voto 1_mod2', 'Voto 2_mod2', 'Voto 3_mod2', 'Voto 4_mod2',
    'Voto 5_mod2', 'Voto 6_mod2', 'Voto 7_mod2',
    'Voto 8_mod2', 'Voto 9_mod2', 'Voto 10_mod2', 'Voto 11_mod2', 'Voto 12_mod2'
]

df_merged = df_merged[col_order]

# ----------------------------------------------------
# FIX: normalize decimal separator (comma â†’ dot)
# ----------------------------------------------------
numeric_cols = df_merged.columns.drop('ID number')

for col in numeric_cols:
    df_merged[col] = (
        df_merged[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .astype(float)
    )

# Fill missing values AFTER conversion
df_merged = df_merged.fillna(0.0)

# Compute total score
total = df_merged['Results_mod1'] + df_merged['Results_mod2']
df_merged.insert(3, 'Total', total)
```

```python
df_merged.to_csv('data/PBS_mod1_mod2.csv')
```
